```python
from transformers import pipeline

# Load the zero-shot classification pipeline with a multilingual model
classifier = pipeline("zero-shot-classification", model="MoritzLaurer/mDeBERTa-v3-base-xnli-multilingual-nli-2mil7")
#paraphrase-multilingual-mpnet-base-v2

# Define the candidate labels in German
#candidate_labels = ["Politik", "Gesundheitsauswirkungen", "Design-Interventionen", "Beschwerden"]
candidate_labels = ["Schulorganisation", "Innenraumklima", "Wasser/Abkühlung", "Allgemeiner Klima/Umweltdiskurs", "Beschwerden", "Sonstiges"]

# Function to classify a single text
def classify_text(text):
    result = classifier(text, candidate_labels)
    return result['labels'][0], result['scores'][0]  # Return top label and confidence

# Apply classification to the 'Responses' column
df[['topic', 'confidence']] = df['Responses'].apply(lambda x: pd.Series(classify_text(x)))

```

```python
# Display the first few rows with classifications
print(df.head(163))
```
